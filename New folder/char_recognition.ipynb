```python
import common_functions as cf
from skimage.morphology import binary_erosion, binary_dilation, binary_closing,skeletonize, thin
```

```python
hi = cf.io.imread('hi_there.png', as_gray = True)
cf.io.imshow(hi)
```

```python
def histogram_projection (image):
    img = binary_dilation(image)
    horizontal_hist = cf.np.sum(img,axis=1,keepdims=True)
    vertical_hist = cf.np.sum(img,axis=0,keepdims=True)

    return horizontal_hist, vertical_hist
```

```python
h_h, v_h = histogram_projection(hi)
# cf.showHist(h_h)
# cf.showHist(v_h)
cf.plt.hist(h_h)
cf.plt.show()
cf.plt.hist(v_h)
cf.plt.show()
```
